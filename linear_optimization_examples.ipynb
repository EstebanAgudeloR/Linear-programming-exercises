```python
## Exercise 1: Max Flow in 8-Node Transportation Network



# variables de decisión xij: cantidad de carros que van del nodo i al nodo j (i=1..7, j=1..7)
x = cvx.Variable((8,8), nonneg = True)

# función objetivo TENER CUIDADO CON VERIFICAR QUE SÍ ESTAMOS TRABAJANDO SOBRE EL NODO DE DESTINO
Z = cvx.Maximize(cvx.sum(x[:,7])) # Maximizamos la cantidad que sale de la ciudad, representandolo como la cantidad que llega al nodo destino

# restricciones

    #  I0  I1  I2  I3  I4  I5  I6  I7        #Flujo máximo que puede salir de la columna j a la fila i
b = [[ 0,  0,  0,  0,  0,  0,  0,  0], #I0
     [ 7,  0,  0,  0,  0,  0,  0,  0], #I1
     [25,  0,  0,  0,  0,  0,  0,  0], #I2
     [5,  0,  0,  0,  0,  0,  0,  0], #I3
     [ 0, 10, 20,  0,  0,  0,  0,  0], #I4
     [ 0,  0,  9,  4,  0,  0,  0,  0], #I5
     [ 0,  0,  0, 15,  0,  0,  0,  0], #I6
     [ 0,  0,  0,  0, 18, 40,  9,  0]] #I7


r = [
    # de capacidad máxima por tramo xij
    x <= b,

    # carros que entra debe ser igual a la que sale
    cvx.sum(x[0,:]) == cvx.sum(x[:,7]),

     # nodos de transbordo
    cvx.sum(x[:,1]) - cvx.sum(x[1,:]) >= 0,
    cvx.sum(x[:,2]) - cvx.sum(x[2,:]) >= 0,
    cvx.sum(x[:,3]) - cvx.sum(x[3,:]) >= 0,
    cvx.sum(x[:,4]) - cvx.sum(x[4,:]) >= 0,
    cvx.sum(x[:,5]) - cvx.sum(x[5,:]) >= 0,
    cvx.sum(x[:,6]) - cvx.sum(x[6,:]) >= 0

]

# Configuración del problema
P = cvx.Problem(Z,r)

# Solucion el problema
P.solve()

print(np.round(P.value,1))
print(np.round(x.value,1))
```
